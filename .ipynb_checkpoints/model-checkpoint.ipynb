### Clustering the Data

```python
#Identifying Optimum Number of Clusters
#Create Fuction to work out optimum number of clusters

def opt_k_means(data, max_k):
    means=[]
    inertias = []
    
    for k in range(1, max_k):
        kmeans= KMeans(n_clusters = k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
        
     #Generate the elbow plot
    fig = plt.subplots(figsize = (10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
         
```

```python
opt_k_means(df[['RHOB.T', 'NPHI.T']], 10)
```

### Applying K Means Clustering

```python
Kmeans = KMeans(n_clusters = 3)
```

```python
Kmeans.fit(df[['NPHI.T', 'RHOB.T']])
```

```python
df['kmeans_3']= Kmeans.labels_
```

### Plotting The Results

```python
plt.scatter(x = df['NPHI'], y = df['RHOB'], c= df['kmeans_3'], cmap='viridis')
plt.xlim(-0.1, 1)
plt.ylim(3, 1.5)
plt.show
```

### Creating Multiple Clusters

```python
for k in range(1,6):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(df[['RHOB.T','NPHI.T']])
    df[f'KMeans_{k}']= kmeans.labels_
```

```python
df
```

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20,5))

for i, ax in enumerate(fig.axes, start=1):
    ax.scatter(x = df['NPHI'], y = df['RHOB'], c = df[f'KMeans_{i}'])
    ax.set_ylim(3, 1.5)
    ax.set_xlim(0, 1)
    ax.set_title(f'N Clusters: {i}')
```
